```python
import requests
from bs4 import BeautifulSoup
def get_links_recursive(base, path, visited, max_depth=5, depth=0):
    print("----------")
    if depth < max_depth:
        try:
            soup = BeautifulSoup(requests.get(base + path).text, "html.parser")
            for link in soup.find_all("a"):
                href = link.get("href")
                if href not in visited:
                    html = request.urlopen(href).read().decode('utf8')
                    raw = BeautifulSoup(html, 'html.parser').find('div').get_text()
                    visited.append(raw)
                    print(f"at depth {depth}: {href} : {(visited)}")
                    if(len(visited)>=5):
                        hello=visited
                        return visited
                    if href.startswith("http"):
                        get_links_recursive(href, "", visited, max_depth, depth + 1)
                    else:
                        get_links_recursive(base, href, visited, max_depth, depth + 1)
        except:
            pass
```
