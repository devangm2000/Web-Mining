```python
import re
from bs4 import BeautifulSoup
from urllib import request
from nltk.corpus import stopwords
from nltk import word_tokenize
import pandas as pd
```

```python
url1 = "https://en.wikipedia.org/wiki/Data_mining"
html1 = request.urlopen(url1).read().decode('utf8')
raw1 = BeautifulSoup(html1, 'html.parser').get_text()
w1 = word_tokenize(raw1)
for i in range(816,5000):
    print(w1[i], end =" ")
```

```python
from nltk import word_tokenize
import itertools
stop_words = stopwords.words('english')
```

```python
setA = set(w1)
setB = set(w2)
setC = set(w3)
setD = set(w4)
setE = set(w5)

totalSet = setA|setB|setC|setD|setE
totalList = list(totalSet)
listOfWords = ['Document ID']
for i in range(5):
    listOfWords.append(totalList[i])

docA =['doc1']
docB =['doc2']
docC =['doc3']
docD =['doc4']
docE =['doc5']

for w in listOfWords:
    if(w == "Document ID"):
        continue
    if (w in w1):
        docA.append(1)
    else:
        docA.append(0)
        
    if (w in w2):
        docB.append(1)
    else:
        docB.append(0)
    
    if (w in w3):
        docC.append(1)
    else:
        docC.append(0)
        
    if (w in w4):
        docD.append(1)
    else:
        docD.append(0)
        
    if (w in w5):
        docE.append(1)
    else:
        docE.append(0)
        
data = [docA,docB,docC,docD,docE]
data
df = pd.DataFrame(data, columns = listOfWords)
df
```

```python
docA=['doc1']
docB=['doc2']
docC=['doc3']
docD=['doc4']
docE=['doc5']

for w in listOfWords:
    if(w == "Document ID"):
        continue
    if (w in w1):
        docA.append(w1.count(w))
    else:
        docA.append(0)
    
    if (w in w2):
        docB.append(w2.count(w))
    else:
        docB.append(0)

    if (w in w3):
        docC.append(w3.count(w))
    else:
        docC.append(0)
    
    if (w in w4):
        docD.append(w4.count(w))
    else:
        docD.append(0)
        
    if (w in w5):
        docE.append(w5.count(w))
    else:
        docE.append(0)

        
data = [docA,docB,docC,docD,docE]
data
df = pd.DataFrame(data, columns = listOfWords)
df
```

```python
filtered1 = [w for w in w1 if not w in stop_words]
filtered2 = [w for w in w2 if not w in stop_words]
filtered3 = [w for w in w3 if not w in stop_words]
filtered4 = [w for w in w4 if not w in stop_words]
filtered5 = [w for w in w5 if not w in stop_words]

file1 = open("A.doc","w", encoding='utf-8')
file2 = open("B.doc","w",encoding='utf-8')
file3 = open("C.doc","w", encoding='utf-8')
file4 = open("D.doc","w",encoding='utf-8')
file5 = open("E.doc","w", encoding='utf-8')

for w in filtered1:
    file1.write(w+" ")
for w in filtered2:
    file2.write(w+" ")
for w in filtered3:
    file3.write(w+" ")
for w in filtered4:
    file4.write(w+" ")
for w in filtered5:
    file5.write(w+" ")
    
print("Successfully saved the filtered words into 5 seperate files!")
print("Successfully saved the filtered words into 5 seperate files!")
```
