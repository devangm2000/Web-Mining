```python
from bs4 import BeautifulSoup
from urllib import request
from nltk.corpus import stopwords
from nltk import word_tokenize
import pandas as pd
url1 = "https://en.wikipedia.org/wiki/Web_mining"
url2 = "https://en.wikipedia.org/wiki/Data_mining"
url3 = "https://en.wikipedia.org/wiki/Artificial_intelligence"
url4 = "https://en.wikipedia.org/wiki/Machine_learning"
url5 = "https://en.wikipedia.org/wiki/Mining"
html1 = request.urlopen(url1).read().decode('utf8')
raw1 = BeautifulSoup(html1, 'html.parser').get_text()
html2 = request.urlopen(url2).read().decode('utf8')
raw2 = BeautifulSoup(html2, 'html.parser').get_text()
html3 = request.urlopen(url3).read().decode('utf8')
raw3 = BeautifulSoup(html3, 'html.parser').get_text()
html4 = request.urlopen(url4).read().decode('utf8')
raw4 = BeautifulSoup(html4, 'html.parser').get_text()
html5 = request.urlopen(url5).read().decode('utf8')
raw5 = BeautifulSoup(html5, 'html.parser').get_text()

bagOfW1 = raw1.lower().split()
bagOfW2 = raw2.lower().split()
bagOfW3 = raw3.lower().split()
bagOfW4 = raw4.lower().split()
bagOfW5 = raw5.lower().split()

bofW1 = []
bofW2 = []
bofW3 = []
bofW4 = []
bofW5 = []

```

```python
for x in bagOfW1:
    if x.isalpha() and x not in stopwords.words("english"):
        bofW1.append(x)
for x in bagOfW2:
    if x.isalpha() and x not in stopwords.words("english"):
        bofW2.append(x)
for x in bagOfW3:
    if x.isalpha() and x not in stopwords.words("english"):
        bofW3.append(x)
for x in bagOfW4:
    if x.isalpha() and x not in stopwords.words("english"):
        bofW4.append(x)
for x in bagOfW5:
    if x.isalpha() and x not in stopwords.words("english"):
        bofW5.append(x)

bagOfWords1 = []
bagOfWords2 = []
bagOfWords3 = []
bagOfWords4 = []
bagOfWords5 = []
```

```python
from nltk import PorterStemmer
stemmer = PorterStemmer()
for x in bofW1:
    temp = stemmer.stem(x)
    if x not in bagOfWords1:
        bagOfWords1.append(temp)
for x in bofW2:
    temp = stemmer.stem(x)
    if x not in bagOfWords2:
        bagOfWords2.append(temp)
for x in bofW3:
    temp = stemmer.stem(x)
    if x not in bagOfWords3:
        bagOfWords3.append(temp)
for x in bofW4:
    temp = stemmer.stem(x)
    if x not in bagOfWords4:
        bagOfWords4.append(temp)
for x in bofW5:
    temp = stemmer.stem(x)
    if x not in bagOfWords5:
        bagOfWords5.append(temp)

d6 = "Mining large volume of data"
bofWq = d6.lower().split()

for x in bofWq:
    temp = stemmer.stem(x)
    if x not in bofWq:
        bofWq.append(temp)
uniqueWordsq = (set(bofWq))
numOfWordsq = dict.fromkeys(uniqueWordsq, 0)
for word in bofWq:
    numOfWordsq[word] += 1
print(numOfWordsq)
tfq = computeTF(numOfWordsq, bofWq)
print(tfq)
idfsq = computeIDF([tfq])
print(idfsq)
tfidfq = computeTFIDF(tfq, idfsq)
dfq = pd.DataFrame([tfidfq])
dfq

uniqueWords1 = set(bagOfWords1).union(set(bagOfWords2)).union(set(bagOfWords3)).union(set(bagOfWords4)).union(set(bagOfWords5))

numOfWords1 = dict.fromkeys(uniqueWords1, 0)
```

```python
for word in bagOfWords1:
    numOfWords1[word] += 1
numOfWords2 = dict.fromkeys(uniqueWords1, 0)
for word in bagOfWords2:
    numOfWords2[word] += 1
numOfWords3 = dict.fromkeys(uniqueWords1, 0)
for word in bagOfWords3:
    numOfWords3[word] += 1
numOfWords4 = dict.fromkeys(uniqueWords1, 0)
for word in bagOfWords4:
    numOfWords4[word] += 1
numOfWords5 = dict.fromkeys(uniqueWords1, 0)
for word in bagOfWords5:
    numOfWords5[word] += 1

def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict

tf1 = computeTF(numOfWords1, bagOfWords1)
tf2 = computeTF(numOfWords2, bagOfWords2)
tf3 = computeTF(numOfWords3, bagOfWords3)
tf4 = computeTF(numOfWords4, bagOfWords4)
tf5 = computeTF(numOfWords5, bagOfWords5)
print(tf1)

```

```python
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N + 1 / float(val))
    return idfDict

idfs = computeIDF([numOfWords1, numOfWords2 , numOfWords3, numOfWords4,numOfWords5])
print(idfs)

def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf

import pandas as pd
tfidf1 = computeTFIDF(tf1, idfs)
tfidf2 = computeTFIDF(tf2, idfs)
tfidf3 = computeTFIDF(tf3, idfs)
tfidf4 = computeTFIDF(tf4, idfs)
tfidf5 = computeTFIDF(tf5, idfs)
df = pd.DataFrame([tfidf1, tfidf2 , tfidf3, tfidf4 , tfidf5])
df

```
