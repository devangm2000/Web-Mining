# D) Maximum, Minimum and Average Degree of the Network.

```python
#Converting to a Dict
dict1=dict(G.degree())
dict1
```

## Maximum Degree:

```python
maximum = max(dict1, key=dict1.get)
print("The node with maximum degree is",maximum,"and has degree=",dict1[maximum])
```

## Minimum Degree:

```python
minimum =min(dict1, key=dict1.get)
print("The node with minimum degree is",minimum,"and has degree=",dict1[minimum])
```

## Average Degree:

```python
arr=[]
for i in dict1.values():
    arr.append(i)
avg=sum(arr)
avg=avg/len(arr)
print("The average degree in the graph:",avg)
```
