```python
#Count the frequency of repetition of each word found in each file.
fopen = open("Artificaial intelligence.txt", "r")
d = dict()
for line in fopen:
    line = line.strip()
    words = line.split(" ")
    for word in words:
        if word in d:
            d[word] = d[word] + 1
        else:
            d[word] = 1
for key in list(d.keys()): 
    print(key, ":", d[key])
```
