```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 45.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
file_content1=open ("doc1.doc").read()
file_content2=open ("doc1.doc").read()
```

```python
wordcloud1 = WordCloud(font_path = r'C:\Windows\Fonts\Verdana.ttf',
                            stopwords = STOPWORDS,
                            background_color = 'white',
                            width = 1200,
                            height = 1000,
                            color_func = random_color_func
                            ).generate(file_content1)
wordcloud2 = WordCloud(font_path = r'C:\Windows\Fonts\Verdana.ttf',
                            stopwords = STOPWORDS,
                            background_color = 'white',
                            width = 1200,
                            height = 1000,
                            color_func = random_color_func
                            ).generate(file_content2)
```

```python
plt.imshow(wordcloud1)
plt.axis('off')
plt.show()
```

```python
plt.imshow(wordcloud2)
plt.axis('off')
plt.show()
```
